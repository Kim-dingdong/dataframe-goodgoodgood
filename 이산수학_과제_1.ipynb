열이랑 다른점은 데이터프레임 이름뒤에 .iloc혹은 loc붙이는게 다르네요



````python
# 새로운 데이터프레임 생성
exam_data = {'수학':[90,80,100], '영어':[98,89,100],
            '음악':[85,95,100], '체육':[100,60,100]}
df = pd.DataFrame(exam_data,index=['호날두','칸예웨스트','일론머스크'])
df


df
````

**데이터프레임에서 행만 뽑아서 보는방법은?**

````python
df_index1 = df.loc['일론머스크']#loc로 일론머스크의 index명
df_position1 = df.iloc[2]#iloc로 일론머스크의 위치
df_index1#이런식으로 일론머스크 행의 자료만 따로 볼 수 있다. df_position1도 동일
````

````python
df_index2= df.loc[['일론머스크','칸예웨스트']]#대괄호 두개쓰면 두개의 index를 동시에 뽑아서 쓸 수 있
df_index2=df.iloc[[2,1]]#위와 동일
df_index2
````
